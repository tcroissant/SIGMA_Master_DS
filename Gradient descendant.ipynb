```python
#function to import and export data from cPickle format

def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding = 'bytes')
    return dict

def save_obj(obj, name):
    with open('export/' + name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

# 3) Descente de gradient

```python
#Fonction de coût. Utilisation d'un fonction Python
```

```python
def cout(a,b):
    #a  matrice des vecteurs de parametre theta 
    #b  matrice des vraies valeurs
    m = b.shape[0] # le nombre d'exemples dans l'echantillon
    return  (np.sum(np.square(a-(b.reshape(-1,1))))/(2*m))
```

```python
m_cout = cout(modele(featbiai,theta),label)
```

```python
print(m_cout)
```

```python
# Dérivée de la fonction de cout == gradients.
# Attention aux dimensions lors de la somme.
# Poser le caltul matriciel sur papier pour être sûr.
```

```python
def d_cout(a,b,c):
    # a matrice des vecteurs de parametre theta, y^
    # b vecteur des valeurs à prédire, y
    # c matrice des variables, x
    m = c.shape[0]
    return ((1/m)*np.dot(c.T,(a-b.reshape((-1,1)))))
```

```python
print(d_cout(modele(featbiai,theta),label,featbiai))
```
